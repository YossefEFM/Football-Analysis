# ***Import Libraries***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/fifa-21-complete-player-dataset/fifa21_male2.csv')
```

# ***Data Cleaning***

```python
# Handle missing values
data = data.dropna()

# Remove duplicates
data = data.drop_duplicates()
```

 # ***Player Performance Analysis***

```python
performance_indicators = ['OVA', 'POT', 'Acceleration', 'Sprint Speed', 'Shot Power']


data[performance_indicators].hist(bins=20, figsize=(12, 8))
plt.show()


top_performers = data.sort_values(by='OVA', ascending=False).head(10)
print(top_performers[['Name', 'OVA']])
```

# ***Positional Analysis***

```python
avg_rating_by_position = data.groupby('Position')['OVA'].mean().sort_values(ascending=False)
print(avg_rating_by_position)
```

# ***Age and Performance***

```python
sns.scatterplot(x='Age', y='OVA', data=data)
plt.show()
```

# ***Club Analysis***

```python
avg_rating_by_club = data.groupby('Club')['OVA'].mean().sort_values(ascending=False)
print(avg_rating_by_club.head(10))
```

# ***Nationality Analysis***

```python
avg_rating_by_nationality = data.groupby('Nationality')['OVA'].mean().sort_values(ascending=False)
print(avg_rating_by_nationality.head(10))
```

# ***Correlation Analysis***

```python
non_numeric_columns = data.select_dtypes(exclude=['float64', 'int64']).columns
plt.figure(figsize=(50, 50))
numeric_df = data.drop(non_numeric_columns, axis=1)
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
# Select 20 top positions from data 
top_positions = data['Position'].value_counts().head(30).index
sns.boxplot(x='Position', y='OVA', data=data[data['Position'].isin(top_positions)])
plt.xticks(rotation=45)
plt.show()

top_clubs = data['Club'].value_counts().head(10)
top_clubs.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Clubs by Number of Players')
plt.xlabel('Club')
plt.ylabel('Number of Players')
plt.show()
```

# ***Insights and Recommendations***

```python
print("Average overall rating by position:\n", avg_rating_by_position)
print("\nTop performers:\n", top_performers[['Name', 'OVA']])
```
